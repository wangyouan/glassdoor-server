```python
from pathlib import Path
import pandas as pd
```

```python
# 数据目录
base = Path("/data/disk4/database/Revelio Labs/revelio_sentiment")

# 两个压缩包
zip_reviews = base / "sentiment_individual_reviews.zip"
zip_scores = base / "sentiment_scores.zip"

def preview_zip_csv(zip_path, sample_nrows=100000, head_n=5):
    print("=" * 100)
    print(f"File: {zip_path.name}")
    
    # 先只读取样本，快速看结构
    df_sample = pd.read_csv(
        zip_path,
        compression="zip",
        nrows=sample_nrows,
        low_memory=False
    )
    
    print(f"Sample shape (nrows={sample_nrows}): {df_sample.shape}")
    print("\nColumns:")
    print(df_sample.columns.tolist())
    
    print("\nDtypes:")
    print(df_sample.dtypes)
    
    print("\nHead:")
    display(df_sample.head(head_n))
    
    print("\nMissing ratio (top 15):")
    miss = df_sample.isna().mean().sort_values(ascending=False).head(15)
    print(miss)

# 1) 大文件：先看样本
preview_zip_csv(zip_reviews, sample_nrows=100000, head_n=5)

# 2) 小文件：可直接全量读
print("\n" + "=" * 100)
print(f"File: {zip_scores.name} (full load)")
df_scores = pd.read_csv(zip_scores, compression="zip", low_memory=False)

print(f"Full shape: {df_scores.shape}")
print("\nColumns:")
print(df_scores.columns.tolist())

print("\nDtypes:")
print(df_scores.dtypes)

print("\nHead:")
display(df_scores.head(10))
```

```python
# sentiment_individual_reviews: review_date 时间范围 + observation 数量
zip_reviews = Path("/data/disk4/database/Revelio Labs/revelio_sentiment/sentiment_individual_reviews.zip")

n_obs = 0
min_date = pd.NaT
max_date = pd.NaT

for chunk in pd.read_csv(
    zip_reviews,
    compression="zip",
    usecols=["review_date"],
    parse_dates=["review_date"],
    chunksize=1_000_000,
    low_memory=False,
):
    n_obs += len(chunk)
    s = chunk["review_date"].dropna()
    if not s.empty:
        cmin = s.min()
        cmax = s.max()
        min_date = cmin if pd.isna(min_date) else min(min_date, cmin)
        max_date = cmax if pd.isna(max_date) else max(max_date, cmax)

print(f"Total observations: {n_obs:,}")
print(f"review_date min: {min_date}")
print(f"review_date max: {max_date}")
```

```python
from pathlib import Path
from zipfile import ZipFile

base_common = Path(
    "/data/disk4/database/Revelio Labs/revelio_common"
)

print(f"Target path: {base_common}")
print(f"Exists: {base_common.exists()}")
print(f"Is dir: {base_common.is_dir()}\n")

if not base_common.exists():
    print("[WARN] Path does not exist in current kernel environment.")
else:
    zip_files = sorted(base_common.rglob("*.zip"))
    print(f"Found {len(zip_files)} zip file(s) recursively under:\n{base_common}\n")

    for zp in zip_files:
        print("=" * 100)
        print(f"ZIP: {zp}")
        try:
            with ZipFile(zp, "r") as zf:
                infos = zf.infolist()
                if not infos:
                    print("  (empty zip)")
                else:
                    print(f"  Contains {len(infos)} item(s):")
                    for info in infos:
                        print(f"  - {info.filename} ({info.file_size:,} bytes)")
        except Exception as e:
            print(f"  [ERROR] Cannot read zip: {e}")
```
